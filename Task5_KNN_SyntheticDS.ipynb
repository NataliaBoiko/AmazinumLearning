```python
import pandas as pd 
import numpy as np 
import ML_mst as mst

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib notebook
from matplotlib.colors import ListedColormap
cmap_bold = ListedColormap(['blue','#FFFF00','black','green'])

np.random.seed= 2021
X_D2, y_D2 = make_blobs(n_samples = 300, n_features = 2, centers = 8,
                       cluster_std = 1.3, random_state = 4)
y_D2 = y_D2 % 2
plt.figure()
plt.title('Sample binary classification problem with non-linearly separable classes')
plt.scatter(X_D2[:,0], X_D2[:,1], c=y_D2,
           marker= 'o', s=30, cmap=cmap_bold)
```

```python
from sklearn.model_selection import train_test_split # sklearn 
X_train, X_test, y_train, y_test = train_test_split(X_D2, y_D2, random_state=0, test_size=0.25, shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier # sklearn 
n_neighbors = 25
knn = KNeighborsClassifier(n_neighbors)
knn.fit(X_train, y_train)
```

```python
k_best = n_neighbors
score_best = knn.score(X_test, y_test)
print ('The best k = {} , score = {}'.format(k_best,score_best ))
```

```python
mst.plot_decision_boundary(knn, X_test, y_test, X_test=X_test, y_test= y_test,
                               title='KNN classification K= {}, score = {:.2f}'.format(k_best, score_best), 
                               precision=0.05,plot_symbol_size = 50)
```
